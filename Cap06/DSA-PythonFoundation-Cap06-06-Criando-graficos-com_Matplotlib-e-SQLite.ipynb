```python
# Remove o arquivo com o banco de dados SQLite (caso exista)
import os
os.remove("dsa.db") if os.path.exists("dsa.db") else None
```

```python
import sqlite3
import random
import datetime
import time
import matplotlib.pyplot as plt
%matplotlib notebook
# matplotlib notebook = criar grafico dentro do jupyter notebook

# Criando uma conexão
conn = sqlite3.connect("dsa.db")

# Criando um cursor
c = conn.cursor()

# Função para criar uma tabela
def create_table():
    c.execute("CREATE TABLE IF NOT EXISTS produtos(id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, date TEXT, \
              prod_name TEXT, valor REAL)")
    
create_table()

# Função para inserir uma linha
def data_insert():
    c.execute("INSERT INTO produtos VALUES(002, '02-05-2016', 'teclado', 130 )")
    conn.commit()
    
data_insert()
    
# Usando varáveis para inserir dados
def data_insert_var():
    new_date = datetime.datetime.now()
    new_prod_name = "monitor"
    new_valor = random.randrange(50,100)
    c.execute("INSERT INTO produtos (date, prod_name, valor) VALUES ( ?, ?, ?)",
             (new_date, new_prod_name, new_valor))
    conn.commit()
    
for i in range(10):
    data_insert_var()
    time.sleep(1)
    
# Leitura de dados
def leitura_todos_dados():
    c.execute("SELECT * FROM PRODUTOS")
    for linha in c.fetchall():
        print(linha)
        

# Leitura de colunas específicas
def leitura_colunas():
    c.execute("SELECT * FROM PRODUTOS")
    for linha in c.fetchall():
        print(linha[3])
        
# Update
def atualiza_dados():
    c.execute("UPDATE produtos SET valor = 70.00 WHERE valor >= 90.0")
    conn.commit
    
# Delete
def remove_dados():
    c.execute("DELETE FROM produtos WHERE valor <= 60.0")
    conn.commit()
    
# Gerar gráfico com os dados no banco de dados
def dados_graficos():
    c.execute("SELECT id, valor FROM produtos")
    ids = []
    valores = []
    dados = c.fetchall()
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])
        
    plt.bar(ids, valores)
    plt.show()
```

```python
atualiza_dados()
```

```python
remove_dados()
```

```python
dados_graficos()
```
